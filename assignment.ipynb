```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
# 辞書型変数
sitemap = {}

# 訪問したところを記録
visited = set()

# これから訪問するURL
to_visit = [('https://www.musashino-u.ac.jp/', 0)]
```

```python
url = 'https://www.musashino-u.ac.jp'
res = requests.get(url)
res.encoding = res.apparent_encoding
print(f"ステータスコード: {res.status_code}")  # ステータスコードを表示
soup = BeautifulSoup(res.text, 'html.parser')
print(f"soupの型: {type(soup)}")  # soupオブジェクトの型を表示
```

```python
# リストが空になるまで繰り返す
while len(to_visit) > 0:
    url, depth = to_visit.pop(0)
    
    # もう訪問していたらスキップ
    if url in visited:
        continue
    
    # 武蔵野大学のサイト以外はスキップ
    if not url.startswith('https://www.musashino-u.ac.jp'):
        continue
    
    if depth > 3:
        continue
    
    visited.add(url)
    
    try:
        print(f"アクセス中: {url}")
        time.sleep(1)
        
        response = requests.get(url)
        
        if response.status_code == 404:
            print(f"  404エラー: スキップ")
            continue
        
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # タイトルを取得
        title_tag = soup.find('title')
        if title_tag:
            title = title_tag.string
        else:
            title = "タイトルなし"
        
        # 辞書に追加
        sitemap[url] = title
        
        # 取得したタイトルを表示
        print(f"  タイトル: {title}")

        # ページ内のリンクを全部取得
        links = soup.find_all('a')
        for link in links:
            href = link.get('href')
            if href:
                # URLを完全な形にする
                if href.startswith('http'):
                    full_url = href
                elif href.startswith('/'):
                    full_url = 'https://www.musashino-u.ac.jp' + href
                else:
                    full_url = 'https://www.musashino-u.ac.jp/' + href
                
                # まだ訪問してないURLをリストに追加
                if full_url not in visited:
                    to_visit.append((full_url, depth + 1))
    
    except:
        sitemap[url] = "エラー"


# 結果を表示
print(sitemap)
```
